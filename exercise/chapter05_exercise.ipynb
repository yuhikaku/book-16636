# 問1

売上高以外にも、気温と最寄り駅距離という２つの変数があります。

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

df = pd.read_excel("data/chapter05_exercise.xlsx")
df.head()
```

# 問2

右上がりの直線が描けているため、正の相関がみられます。

```python
sns.regplot(
    data=df,
    x="気温",
    y="売上高",
    ci=None,
    line_kws={"color": "red"}
)
plt.grid()
```

# 問3

それぞれa=0.26, b=0.006, c=7.35となります。

気温と売上高の関係性を直線で近似すると、気温が1度上昇すると売上高は0.26増えると予想できます。

もし売上高の単位が100万円であれば、26万円増えることになります。

```python
z = df[["気温", "最寄り駅距離"]]
y = df[["売上高"]]

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(z, y)

print(model.coef_)
print(model.intercept_)
```

# 問4

28.7％です。売上高の約3分の1が気温と最寄駅からの距離で説明できることになります。

```python
print(model.score(z, y))
```
