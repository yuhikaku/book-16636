```python
from pathlib import Path
text = Path("data/chapter14_exercise.txt").read_text("utf-8")
print(text)
```

```python
import MeCab
import unidic
import collections
from wordcloud import WordCloud
```

```python
tagger = MeCab.Tagger()

node = tagger.parseToNode(text)

words = []
while node:
    node_features = node.feature.split(",")
    if node_features[0] == "名詞":
        words.append(node.surface)
    node = node.next
```

```python
w_file = " ".join(words)
```

```python
import matplotlib.pyplot as plt

wordcloud = WordCloud(
    font_path="C:/Windows/Fonts/MSGOTHIC.TTC",
    background_color="black",
    colormap="summer",
    collocations=False,
    regexp=r"[^ ]+",
)
wordcloud.generate(w_file)

plt.imshow(wordcloud)
plt.axis("off")
```
