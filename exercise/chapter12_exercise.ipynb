# 問1

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

S = 0.95
I = 0.05
R = 0

inf = 0.014
cont = 0.018
num = 500

S_list, I_list, R_list = [], [], []

for i in range(num):
    S_t1 = S + (-1 * inf * S * I)
    I_t1 = I + (inf * S * I - cont * I)
    R_t1 = R + (cont * I)

    S_list.append(S_t1)
    I_list.append(I_t1)
    R_list.append(R_t1)

    S = S_t1
    I = I_t1
    R = R_t1
```

```python
fig = plt.figure()
plt.plot(S_list, label="S")
plt.plot(I_list, label="I")
plt.plot(R_list, label="R")

plt.title("契約の普及")
plt.xlabel("時間")
plt.ylabel("未契約率or契約率")
plt.legend()
plt.show()
```

# 問3

口コミが未契約者に与える影響力と契約定着率に注目してみましょう。ここでは、$0.014<0.018$、12章では、$0.046>0.018$です。口コミの影響力が契約定着率よりも大きいか小さいかで違いがありそうです。

ちなみに、SIRモデルで、口コミの影響力÷契約定着率に相当するものは**基本再生産数**と呼ばれ、1より大きければ感染が急速に拡大し、小さければ感染は蔓延しません。流行するかどうかの分かれ目です。

ここでは、口コミの影響力÷契約定着率が、1より大きければ普及が進み、小さければ普及は進みません。

参考までに、口コミの影響力÷契約定着率=1を試してみました。500日後の未契約者の割合は74%になります。図に描いてみると、未契約者の割合は高止まりしたままで、本章のように急速な普及が見られません。



```python
# 口コミの影響力÷契約定着率＝１の場合 
S = 0.95
I = 0.05
R = 0

inf = 0.018
cont = 0.018
num = 500

S_list, I_list, R_list = [], [], []

for i in range(num):
    S_t1 = S + (-1 * inf * S * I)
    I_t1 = I + (inf * S * I - cont * I)
    R_t1 = R + (cont * I)

    S_list.append(S_t1)
    I_list.append(I_t1)
    R_list.append(R_t1)

    S = S_t1
    I = I_t1
    R = R_t1
```

```python
fig = plt.figure()
plt.plot(S_list, label="S")
plt.plot(I_list, label="I")
plt.plot(R_list, label="R")

plt.title("契約の普及")
plt.xlabel("時間")
plt.ylabel("未契約率or契約率")
plt.legend()
plt.show()
```
