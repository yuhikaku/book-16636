```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

S = 0.95
I = 0.05
R = 0

inf = 0.014
cont = 0.018
num = 500

S_list, I_list, R_list = [], [], []

for i in range(num):
    S_t1 = S + (-1 * inf * S * I)
    I_t1 = I + (inf * S * I - cont * I)
    R_t1 = R + (cont * I)

    S_list.append(S_t1)
    I_list.append(I_t1)
    R_list.append(R_t1)

    S = S_t1
    I = I_t1
    R = R_t1
```

```python
fig = plt.figure()
plt.plot(S_list, label="S")
plt.plot(I_list, label="I")
plt.plot(R_list, label="R")

plt.title("Â•ëÁ¥Ñ„ÅÆÊôÆÂèä")
plt.xlabel("ÊôÇÈñì")
plt.ylabel("Êú™Â•ëÁ¥ÑÁéáorÂ•ëÁ¥ÑÁéá")
plt.legend()
plt.show()
```
