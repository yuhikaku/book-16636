いえません。売上高を横軸、気温を縦軸にして図を描くと、横軸の売上高が原因で縦軸の気温が結果のように思うかもしれませんが、そんなことはありません。あくまで相関関係であって、因果関係ではありません。

実際、両者の関係は、売上高を縦軸、気温を横軸にして図を描いても同じです。常識の範囲内で考えれば、気温が下がると温かいものが食べたくなり、おでんの売上が伸びると考えるのが妥当でしょう。

おでんと気温の話では、おでんの売上を伸ばせば気温が下がると考える人は少ないでしょうが、相関関係と因果関係を取り違えた議論は意外と身の回りで見受けます。注意が必要です。


```python
df.plot(
    kind="scatter",
    x="気温",
    y="売上高",
    c="blue",
    title="気温と売上高の関係",
    xlabel="気温（℃）",
    ylabel="売上高（万円）"
)
```

```python
print(
    "気温と売上高の関係",
    df.corr()["売上高"]["気温"]
)
```
