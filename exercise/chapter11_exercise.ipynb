災害によって、後に示す図のような状況を考えることになります。
このとき、最短経路はA→C→G→F→Hで、所要時間は23時間です。

```python
import networkx as nx

G = nx.read_weighted_edgelist("data/chapter11_exercise.txt")

path = nx.dijkstra_path(G, "A", "H")
length = nx.dijkstra_path_length(G, "A", "H")

print(path)
print(length)
```

図の描き方は以下の通り。10章のものを再利用したのですが、`#` を付けるとその行( `#G.add_edge("E", "G", weight=2)`)は読まれません。

```python
import networkx as nx
import matplotlib.pyplot as plt

G2 = nx.Graph()

G2.add_node("A")
G2.add_node("B")
G2.add_node("C")
G2.add_node("D")
G2.add_node("E")
G2.add_node("F")
G2.add_node("G")
G2.add_node("H")

G2.add_edge("A", "B", weight=6)
G2.add_edge("A", "C", weight=7)
G2.add_edge("B", "C", weight=3)
G2.add_edge("B", "D", weight=9)
G2.add_edge("B", "E", weight=5)
G2.add_edge("C", "G", weight=8)
G2.add_edge("D", "E", weight=3)
G2.add_edge("D", "F", weight=4)
# G2.add_edge("E", "G", weight=2)
G2.add_edge("F", "G", weight=2)
G2.add_edge("F", "H", weight=6)
G2.add_edge("G", "H", weight=10)
```

```python
pos = nx.spring_layout(G2, seed=7)
nx.draw_networkx_nodes(G2, pos)
nx.draw_networkx_edges(G2, pos)
nx.draw_networkx_labels(G2, pos)

edge_labels = nx.get_edge_attributes(G2, "weight")
nx.draw_networkx_edge_labels(G2, pos, edge_labels)

plt.show()
```
