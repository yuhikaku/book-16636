# 問2

気温の平均は20°C、中央値は18.7°C、標準偏差は6.5、売上高の平均は56万円、中央値は58.8万円、標準偏差は20.5です。

気温の平均は中央値よりわずかに大きく、売上高の平均は中央値よりわずかに小さくなっています。


```python
print("気温", df["気温"].mean())
print("売上高", df["売上高"].mean())

print("気温", df["気温"].median())
print("売上高", df["売上高"].median())

print("気温", df["気温"].std())
print("売上高", df["売上高"].std())
```

## 例3

データフレームから `売上高`を抜き出し、`sns.histplot` を使って描いた場合です。縦軸は密度です。長方形の面積の合計が1になります。

## 例4（おまけ）

例3の解答に、カーネル密度推定（kde：kernel density estimation）を加えました。
カーネル密度推定とは、データから全体の分布を推定した滑らかな曲線です（たとえば、今回のデータが一部の店舗のもので、全国の店舗のデータを推定するイメージです。データ数が増えると滑らかな曲線になります）。

```python
sns.histplot(df["売上高"], kde=True, stat="density")
```

たまにある質問：


> `sns.distplot(df["売上高"])` とすれば、縦軸が密度で、カーネル密度推定もついたヒストグラムが描けるとほかの書籍で勉強したのですが、UserWarningが出ます。どうしたよいでしょう？


`sns.histplot(df["売上高"], kde=True, stat = "density")` で同様なものが描けます。`sns.distplot`はいずれ使えなくなったようなので、今後は、こちらを使うとよいでしょう。
古い（と言っても1～2年前に出版した）書籍のコードを使うと、もうすぐ使えなくなりますというFuture Warningなどが出ることもあります。Warning（警告）が出るとびっくりしますが落ち着いて対処しましょう。もし、エラーメッセージに`deprecated`というワードがあれば、コードが古くなったことがわかります。Pythonは日々更新されているため、コードもアップデートしていく必要があります。
