```python
from pathlib import Path
text = Path("data/chapter14.txt").read_text("utf-8")
print(text)
```

```python
import MeCab
import unidic
import collections
from wordcloud import WordCloud
```

```python
# 次のセルを実行してエラーになる場合はこのコマンドを実行しましょう
# python -m unidic download
```

```python
tagger = MeCab.Tagger()

node = tagger.parseToNode(text)

words = []
while node:
    node_features = node.feature.split(",")
    if node_features[0] == "名詞":
        words.append(node.surface)
    node = node.next
```

```python
words[0:10]
```

```python
ct = collections.Counter(words)
print(ct)
```

```python
w_file = " ".join(words)
w_file
```

```python
import matplotlib.pyplot as plt

wordcloud = WordCloud(
    font_path="C:/Windows/Fonts/MSGOTHIC.TTC",
    background_color="white",
    colormap="tab10",
    collocations=False,
    regexp=r"[^ ]+",
)
wordcloud.generate(w_file)

plt.imshow(wordcloud)
plt.axis("off")
```
