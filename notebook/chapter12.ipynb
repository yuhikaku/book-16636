```python
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
```

```python
S = 0.95
I = 0.05
R = 0

inf = 0.046
cont = 0.018
num = 500

S_list, I_list, R_list = [], [], []

for i in range(num):
    S_t1 = S + (-1 * inf * S * I)
    I_t1 = I + (inf * S * I - cont * I)
    R_t1 = R + (cont * I)

    S_list.append(S_t1)
    I_list.append(I_t1)
    R_list.append(R_t1)

    S = S_t1
    I = I_t1
    R = R_t1
```

```python
fig = plt.figure()
plt.plot(S_list, label="S")
plt.plot(I_list, label="I")
plt.plot(R_list, label="R")

plt.title("契約の普及")
plt.xlabel("時間")
plt.ylabel("未契約率or契約率")
plt.legend()
plt.show()
```
