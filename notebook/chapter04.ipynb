```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

df = pd.read_excel("data/chapter04.xlsx")
df.head()
```

```python
len(df)
```

```python
df.plot(
    kind="scatter",
    x="広告費",
    y="売上高",
    c="blue",
    title="広告費と売上高の関係",
    grid=True
)
```

```python
df.plot(
    kind="scatter",
    x="距離",
    y="売上高",
    c="blue",
    title="最寄駅からの距離と売上高の関係",
    grid=True
)
```

```python
df.plot(
    kind="scatter",
    x="広告費",
    y="距離",
    c="blue",
    title="広告費と距離の関係",
    grid=True
)
```

```python
print(
    "広告費と売上高の関係",
    df.corr()["広告費"]["売上高"]
)
```

```python
print(
    "最寄駅からの距離と売上高の関係",
    df.corr()["距離"]["売上高"]
)
```

```python
print(
    "最寄駅からの距離と広告費の関係",
    df.corr()["距離"]["広告費"]
)
```

```python
print(
    "広告費と広告費の関係",
    df.corr()["広告費"]["広告費"]
)
```

```python
df.corr()
```

```python
sns.heatmap(
    df.corr(),
    annot=True,
    vmax=1,
    vmin=-1,
    center=0,
)
```

```python
sns.heatmap(
    df.corr(),
    annot=False,
    vmax=1,
    vmin=-1,
    center=0,
)
```

```python
# グレースケールで出力する場合
sns.heatmap(
    df.corr(),
    annot=True,
    vmax=1,
    vmin=-1,
    center=0,
    cmap=sns.color_palette("Greys", as_cmap=True)
)
```
